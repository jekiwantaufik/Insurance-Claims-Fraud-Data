```python
import pandas as pd

# Google Sheets links
employee_data_url = 'https://docs.google.com/spreadsheets/d/1gsGhw705fmYvptSXlF3uu2Ei44hSLf8EqKALzHYA_yU/edit?usp=sharing' # Replace with actual link
insurance_data_url = 'https://docs.google.com/spreadsheets/d/1P7_IP-OADle_NpdJxCHW9_H0mlQb2PXGYrkpowOI0U0/edit?usp=sharing' # Replace with actual link
vendor_data_url = 'https://docs.google.com/spreadsheets/d/1YIWfM0G0lBiFwShmI0kddiNBUQQpl1or5RMAC7otbVE/edit?usp=sharing' # Replace with actual link


# Function to read data from Google Sheets
def read_google_sheet(url):
  try:
    # Extract the sheet ID from the URL
    sheet_id = url.split('/d/')[1].split('/edit')[0]
    # Create the export URL
    export_url = f'https://docs.google.com/spreadsheets/d/{sheet_id}/export?format=csv'
    df = pd.read_csv(export_url)
    return df
  except Exception as e:
    print(f"Error reading Google Sheet from {url}: {e}")
    return None

# Load the datasets
employee_data = read_google_sheet(employee_data_url)
insurance_data = read_google_sheet(insurance_data_url)
vendor_data = read_google_sheet(vendor_data_url)

# Display the first few rows of each dataframe
if employee_data is not None:
    print("Employee Data:")
    display(employee_data.head())

if insurance_data is not None:
    print("\nInsurance Data:")
    display(insurance_data.head())

if vendor_data is not None:
    print("\nVendor Data:")
    display(vendor_data.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")

# 1. Count plot of 'STATE' in employee_data
plt.figure(figsize=(12, 6))
sns.countplot(data=employee_data, y='STATE', order=employee_data['STATE'].value_counts().index, palette='viridis')
plt.title('Distribution of Employees Across States')
plt.xlabel('Number of Employees')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

```python
# 3. Bar plot showing the average 'CLAIM_AMOUNT' by 'INSURANCE_TYPE' in insurance_data
plt.figure(figsize=(10, 6))
sns.barplot(data=insurance_data, x='INSURANCE_TYPE', y='CLAIM_AMOUNT', palette='viridis')
plt.title('Average Claim Amount by Insurance Type')
plt.xlabel('Insurance Type')
plt.ylabel('Average Claim Amount')
plt.tight_layout()
plt.show()
```

## Actionable Insights from Exploratory Data Analysis

Based on the comprehensive Exploratory Data Analysis performed on the Employee, Insurance, and Vendor datasets, the following key insights and recommended actions have been identified:

### Data Quality and Missing Values

- **Observation:** Significant missing values were found in `ADDRESS_LINE2` across all datasets, and in `VENDOR_ID`, `AUTHORITY_CONTACTED`, and `CUSTOMER_EDUCATION_LEVEL` in the `Insurance Data`. Missing values in `CITY` were also noted in all datasets, although to a lesser extent.
- **Actionable Insight:** Address missing values through appropriate data imputation techniques or by considering the impact of removal, depending on the downstream analysis or modeling tasks. Prioritize addressing missing `VENDOR_ID` and `AUTHORITY_CONTACTED` in the `Insurance Data` due to their potential importance in fraud detection.

### Data Distribution and Patterns

- **Observation:** The distribution of employees and vendors across states varies, with some states having a higher concentration. The `Insurance Data` reveals the most frequent insurance types and incident severities.
- **Actionable Insight:** Further investigate the factors contributing to the uneven distribution of employees and vendors. Analyze if there are geographical patterns in insurance claims or fraud incidents.

### Numerical Feature Analysis and Anomalies

- **Observation:** Box plots of numerical features in the `Insurance Data` (e.g., `CLAIM_AMOUNT`) indicate the presence of outliers. Scatter plots revealed some patterns, but no strong linear correlations were observed between `PREMIUM_AMOUNT`, `CLAIM_AMOUNT`, `AGE`, and `TENURE`.
- **Actionable Insight:** Develop strategies to handle outliers in numerical features, such as capping or transformation, if necessary for modeling. While no strong linear correlations were found, explore non-linear relationships or interactions between features that might be relevant for fraud detection.

### Categorical Feature Analysis and Infrequent Categories

- **Observation:** Several categorical columns contain infrequent categories, particularly in address-related fields and unique identifiers.
- **Actionable Insight:** Consider grouping or encoding infrequent categories appropriately for modeling to avoid potential issues with low-frequency data. Investigate if specific infrequent categories are associated with a higher risk of fraudulent activities.

### Time Series Analysis

- **Observation:** Time series plots of transaction, policy effective, loss, and report dates show variations in counts over time, with some periods exhibiting higher activity.
- **Actionable Insight:** Analyze these temporal patterns for seasonality or trends that might indicate fraudulent behavior. Consider incorporating time-based features into fraud detection models.

### Vendor and Agent Analysis

- **Observation:** The `Insurance Data` includes `AGENT_ID` and `VENDOR_ID`. There are missing values in `VENDOR_ID`.
- **Actionable Insight:** Investigate the relationships between agents, vendors, and claim characteristics. Analyze if certain agents or vendors are associated with a disproportionate number of claims or potentially fraudulent activities. Address the missing `VENDOR_ID` values if this feature is deemed important for analysis.

### Overall Recommendations

- **Integrate Datasets:** While initial EDA was performed on individual datasets, a more comprehensive understanding can be gained by integrating the datasets based on common identifiers (e.g., `AGENT_ID`, `VENDOR_ID`) to create a unified view for fraud analysis.
- **Feature Engineering:** Create new features based on the insights gained, such as age of policy at the time of loss, duration between loss and report dates, or aggregated claim statistics by agent or vendor.
- **Targeted Fraud Analysis:** Based on the identified anomalies and patterns, focus deeper analysis on specific segments of the data that exhibit higher risk indicators.

This EDA provides a strong foundation for further investigation and the development of a robust insurance claims fraud detection system. The next steps would involve data cleaning and preprocessing based on these insights, followed by feature engineering, model selection, and evaluation.
